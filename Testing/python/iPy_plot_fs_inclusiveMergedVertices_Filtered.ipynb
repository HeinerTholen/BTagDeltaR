# First look TTbar dileptonic

```python
import ROOT
import varial
from rootpy.plotting import Hist
from varial.wrappers import HistoWrapper
```

```python
def stack_vertexDR(fs):
    fs.VertexDRNoMatch.SetFillColor(ROOT.kGreen)
    fs.VertexDROneMatch.SetFillColor(ROOT.kRed)
    fs.VertexDRTwoMatch.SetFillColor(ROOT.kBlue)
    stk = varial.operations.stack((
        HistoWrapper(fs.VertexDRNoMatch),
        HistoWrapper(fs.VertexDROneMatch),
        HistoWrapper(fs.VertexDRTwoMatch),
    ))
    c1 = ROOT.TCanvas()
    stk.stack.Draw("hist")
    c1.BuildLegend()
    return c1
```

# inclusiveMergedVertices collection
### Number of reconstructed IVF vertices and vertex-DR for events with 2 IVF vertices

```python
fs = varial.diskio.read("../fs_ivf_merged.info")
fs.NumIvfVertices
```

```python
stack_vertexDR(fs)
```
